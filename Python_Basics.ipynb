# Core Data structures (Used in ML)


### Lists

```python

## Lists(Pythons ArrayList + More)
nums = [1,2,3,4]

print(nums.append(5)) ##Append
print(nums.pop())     # remove last element
print(nums[0])       # access 1st element
print(nums[-1])      # access last element

```

```python

# Slicing
print(nums[1:3])    # [2,3]
print(nums[::-1])   # reverse
```

```python

# Iteration (don't index unless needed)

for x in nums:
  print(x)


for i,x in enumerate(nums):
  print(i,x)
```
